# Create ML

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier

```

## Data Splitting

```python
X = df.drop(columns=['DEATH_EVENT'])
y = df.DEATH_EVENT

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=42)
```

## Preprocessor

```python
numerical_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="mean")),
    ('scaler',MinMaxScaler())
])

categorical_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('onehot',OneHotEncoder())
])
```

```python
df.columns[:-1]
```

```python
preprocessor = ColumnTransformer([
    ('numeric', numerical_pipeline,['age','creatinine_phosphokinase', 'ejection_fraction', 'platelets','serum_creatinine', 'serum_sodium',  'time'] ),
    ('categoric', categorical_pipeline,['anaemia','diabetes','high_blood_pressure','sex','smoking'])
])
```

```python
X.head()
```

## Pipeline 

```python
pipeline = Pipeline([
    ('preprocessor',preprocessor),
    ('algoritma',KNeighborsClassifier())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_test,y_test), pipeline.score(X_train,y_train)
```

`skor tanpa tuning sudah sangat baik`

## GridSearchCV 

```python
paramater = {
    "algoritma__n_neighbors" : range(1,51,2),
    "algoritma__weights" : ["uniform","distance"],
    "algoritma__p":[1,2]
}

model = GridSearchCV(pipeline, paramater, cv=3, n_jobs=-1 ,scoring='accuracy', verbose=1)
model.fit(X_train,y_train)
```

```python
pd.DataFrame(model.cv_results_).sort_values("rank_test_score")
```

```python
model.best_params_
```

```python
model.score(X_train,y_train), model.score(X_test,y_test)
```
